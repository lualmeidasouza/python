```
# 1) Construindo um gráfico de linha

import matplotlib.pyplot as plt

# Lista de valores para x e y
x=[1,2]
y=[2,3]

# Plota os eixos x e y o gráfico
plt.plot(x,y)

# Mostra o gráfico
plt.show()
```

```
# 2) Construindo legendas para o gráfico 

import matplotlib.pyplot as plt

# Lista de valores para x e y
x=[1,2,5]
y=[2,3,7]

# Título
plt.title("Meu primeiro gráfico com Python")

# Eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

# Plota os eixos x e y o gráfico de linha
plt.plot(x,y)

plt.show()
```

```
# 3) Construindo um gráfico de barras

import matplotlib.pyplot as plt

# Lista de valores para x e y
x=[1,2,3,4,5]
y=[2,3,7,1,0]

titulo = "Gráfico de barras"
eixox = "Eixo X"
eixoy = "Eixo Y"

# Legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

# Plota os eixos x e y o gráfico de barra
plt.bar(x,y)

plt.show()
```

```
# 4) Comparando gráfico de barras

import matplotlib.pyplot as plt

# Lista de valores para x e y
x1=[1,3,5,7,9]
y1=[2,3,7,1,0]

x2=[2,4,6,8,10]
y2=[5,1,3,7,9]


titulo = "Gráfico de barras 2"
eixox = "Eixo X"
eixoy = "Eixo Y"

# Legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

# Plota os eixos x e y o gráfico de barra
plt.bar(x1,y1, label = "Grupo 1")
plt.bar(x2,y2, label = "Grupo 2")
plt.legend()

plt.show()
```

```
# 5) Scatterplot: gráfico de dispersão ou pontos

import matplotlib.pyplot as plt

# Lista de valores para x e y
x1=[1,3,5,7,9]
y1=[2,3,7,1,0]

x2=[2,4,6,8,10]
y2=[5,1,3,7,9]


titulo = "Gráfico de barras 2"
eixox = "Eixo X"
eixoy = "Eixo Y"

# Legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

# Plota os eixos x e y o gráfico de dispersão ou pontos (scatterplot)
plt.scatter(x,y)

plt.show()
```
